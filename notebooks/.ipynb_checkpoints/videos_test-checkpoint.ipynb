# Diffusion config

```python
from easydict import EasyDict as edict 

config = edict()
config.DATASET = "CelebA_HQ" # "LFW" "CelebA_HQ", "FFHQ"
config.batch_size = 5

config.diffusion_name = "controlnet_segmentation_masks" # controlnet_landmarks_5, controlnet_landmarks_26, controlnet_segmentation_masks, ffhq, multi_controlnet
config.guidance = False # default = False
config.guidance_type = "segmentation" # "landmarks", "segmentation"
config.controlnet = True # default = True
config.ddim_inversion = False # default = False
config.evaluation = "jpeg_compression" # "standard", "diversity" "diff_cryptanalysis" "jpeg_compression"

if config.guidance == True:
    config.guidance_scale = 5
else : 
    config.guidance_scale = None

config.fr_models = ["facenet_vggface2", "facenet_casia"]

config.mask = "anonymized_face" # "original_face", "anonymized_face"
config.seeds = "identity"

if config.ddim_inversion:
    prefix = "ddim_inversion"
else :
    prefix = "controlnet_inversion"

if config.guidance == True:
    config.save_path_suffix = f"conditionned_{config.guidance_type}{config.guidance_scale}_mask_{config.mask}_{prefix}"
    # config.save_path_suffix = f"test"
elif config.controlnet == True:
    config.save_path_suffix = f"{config.diffusion_name}_mask_{config.mask}_{prefix}"
else :
    config.save_path_suffix = f"unconditionned_mask_{config.mask}_{prefix}"

if config.evaluation == "diversity":
    config.save_path_suffix = os.path.join("diversity_evaluation", config.save_path_suffix)
    config.n_size = 1000
elif config.evaluation == "standard":
    config.n_size = 2500
elif config.evaluation == "diff_cryptanalysis":
    config.save_path_suffix = os.path.join("diff_cryptanalysis", config.save_path_suffix)
    config.n_size = 1000
elif config.evaluation == "jpeg_compression":
    config.save_path_suffix = os.path.join("jpeg_compression", config.save_path_suffix)
    config.n_size = 1000
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

whereIam = platform.node()
print(whereIam)
if whereIam == "MSI":
    root = "C:\\Users\\Pol\\Documents\\code\\anonymization"
else:
    root = "/lustre/fswork/projects/rech/irm/uzq68by/code/anonymization"


# Load the diffusion model and anonymizer
diffusion_name = config.diffusion_name
diffusion_model = DiffusionModel(name=diffusion_name,
                                inverse_scheduler="hugging_face", torch_device=device)
diffusion_model.to_device(device)
face_parser = BiSeNet(num_classes=19, backbone_name="resnet34")
face_parser.load_state_dict(torch.load(os.path.join(root.split("code")[0], "models", "bisenet", "resnet34.pt")
                                        ,map_location=device))
face_parser = face_parser.to(device).eval()

mtcnn = get_face_detector(detector_name="mtcnn", device="cuda", input_size=256)

flip_anonymizer = FlipAnonymizer(diffusion_model, face_parser, mtcnn, config.guidance, config.guidance_type,
                                    config.guidance_scale, config.controlnet, config.ddim_inversion)
del diffusion_model
```

```python
# Load dataset (synthetic or real)
tfms = transforms.Compose([
            transforms.Resize((512,512)),
            transforms.Normalize(mean=[0.5], std=[0.5])  # normalize grayscale or RGB
        ])
resize = transforms.Compose([transforms.Resize((256,256)),
        transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.5, 0.5))])

retina_net = get_face_detector(detector_name="retinaface", device="cuda", network="resnet50", image_size = [512,512], crop_size=512, margin=150)

# Initialize face recognition pipelines
face_recognizer_models = ["arcface_r50", "facenet_vggface2", "facenet_casia", "adaface_vit"]
fr_pipelines = {model: FRPipeline(face_detector_name="mtcnn", face_recognizer_name=model, device="cuda") for model in face_recognizer_models}

```

```python
import cv2
import torch

frames = ano_video  # list of tensors (C,H,W) in range [0,1]

h, w = frames[0].squeeze(0).shape[1:]
fps = 25

video = cv2.VideoWriter(
    "ano.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (w, h)
)

for frame in frames:
    frame = frame.squeeze(0).permute(1, 2, 0).cpu().numpy()  # HWC
    frame = (frame.clip(-1,1) + 1)/2
    frame = (frame * 255).astype("uint8")
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    video.write(frame)

video.release()
```

```python
import cv2
import torch

frames = recover_video  # list of tensors (C,H,W) in range [0,1]

h, w = frames[0].squeeze(0).shape[1:]
fps = 25

video = cv2.VideoWriter(
    "rec.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (w, h)
)

for frame in frames:
    frame = frame.squeeze(0).permute(1, 2, 0).cpu().numpy()  # HWC
    frame = (frame.clip(-1,1) + 1)/2
    frame = (frame * 255).astype("uint8")
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    video.write(frame)

video.release()
```

```python
################################################################################
############################## Results saving ##################################
################################################################################

for key, value in utility_maintenance_metrics.items():
    for subkey, subvalue in value.items():
        utility_maintenance_metrics[key][subkey] = torch.from_numpy(np.concatenate(subvalue, axis=0))

# Concatenate all templates for each model
final_templates = {}
final_ano_templates = {}
final_recon_templates = {}
results = {'keys': torch.cat(all_keys, dim=0),
            'mae': mae,
            'ssim': ssim,
            'psnr': psnr,
            'fid': fid.compute().item(),
            'fid_recon': fid_recon.compute().item(),
            'utility_maintenance_metrics': utility_maintenance_metrics
            }

for model_name in templates.keys():
    if templates[model_name]:  # Check if we have any templates
        results[model_name + '_templates'] = {}

        final_templates[model_name] = torch.cat(templates[model_name]['clean'], dim=0)
        final_ano_templates[model_name] = torch.cat(templates[model_name]['anonymized'], dim=0)
        final_recon_templates[model_name] = torch.cat(templates[model_name]['reconstructed'], dim=0)

        print(f"Final template shape for {model_name}: {final_templates[model_name].shape}")
        print(f"Final ano template shape for {model_name}: {final_ano_templates[model_name].shape}")
        print(f"Final recon template shape for {model_name}: {final_recon_templates[model_name].shape}")

        results[model_name + '_templates']['clean'] = final_templates[model_name]
        results[model_name + '_templates']['anonymized'] = final_ano_templates[model_name]
        results[model_name + '_templates']['reconstructed'] = final_recon_templates[model_name]

        os.makedirs(f"results/{DATASET}/{config.save_path_suffix}", exist_ok=True)
        torch.save(results, f"results/{DATASET}/{config.save_path_suffix}/results_part_{argparse.n}.pt")
        # torch.save({
        #     'templates': final_templates[model_name],
        #     'templates_ano': final_ano_templates[model_name],
        #     'templates_recon': final_recon_templates[model_name],
        #     'keys': torch.cat(all_keys, dim=0)
        # }, f"results/{DATASET}/{config.save_path_suffix}/{DATASET}_{model_name}_{argparse.n}_templates.pt")

print("Results saved!")
```
